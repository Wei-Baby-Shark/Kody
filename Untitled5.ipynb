```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt

# 函數從網址加載圖片
def load_image(url):
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

# 定義變換
transforms_dict = {
    "Original Image": None,
    "Resized": transforms.Resize((960, 1024)),
    "Center Cropped": transforms.CenterCrop(100),
    "Color Jittered": transforms.ColorJitter(brightness=0.6, contrast=0.7, saturation=0.8),
    "Horizontally Flipped": transforms.RandomHorizontalFlip(p=0.9),
    "Randomly Rotated": transforms.RandomRotation(100)
}

# 加載圖片
img_url = 'https://p2.bahamut.com.tw/B/2KU/90/03a732d87504069f280a19f5481kejq5.JPG?v=1676020474707'
img = load_image(img_url)

# 準備繪圖
fig, ax = plt.subplots(1, 6, figsize=(15, 5))  # 1 row，6 column

# 繪製和顯示變換後的圖像
for i, (title, transform) in enumerate(transforms_dict.items()):
    transformed_img = img
    if transform:
        transformed_img = transform(transformed_img)
    ax[i].imshow(transformed_img)
    ax[i].set_title(title)
    ax[i].axis('off')

plt.tight_layout()
plt.show()
```
